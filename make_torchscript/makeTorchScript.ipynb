## object tracking

```python
from ultralytics import YOLO
import torch

import os

dir_save = r"..\ros_ws\src\tracker_pkg\config"
os.makedirs(dir_save, exist_ok=True)
DEVICE = torch.device("cpu") #torch.device("cuda" if torch.cuda.is_available() else "cpu")
#yolov10

with torch.no_grad():
    model = YOLO("yolov10m.pt")
    model.to(DEVICE)
    # Export the model and save to dir_save
    export_path = os.path.join(dir_save, "yolov10m.torchscript.pt")
    model.export(format="torchscript", imgsz=(480,640), device=DEVICE)#, path=export_path) #(B,C,H,W)
```
